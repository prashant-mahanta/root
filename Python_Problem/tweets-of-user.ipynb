## Streaming user's tweets
Tweepy’s Cursor allows us to stream data from the query without having to manually query for more data in batches.

```python
# array to store all the tweets
tweets = []

# open json file 
fp = open("python-task.jsonl","w+")

# can iterate over user's tweet using user_timeline keep tweet_mode as extended and to get full text of a tweet
for status in tweepy.Cursor(api.user_timeline, id = user_name, tweet_mode = "extended").items():
    tweet = {}
    
    # If a tweet is retweet then extract the favorite count from the tweet from which it is retweeted
    try:
        favorite = status.retweeted_status.favorite_count
    except AttributeError:
        favorite = status.favorite_count
    
    # Whether tweet contains media files(photos)
    try:
        images_count = len([medium['type'] == 'photo' for medium in status.entities['media']]) 
    except:
        
        # If no photo present in tweet return None
        images_count = None
    
    # Full Text of the tweet as tweet_mode is extended
    tweet["text"] = status.full_text
    
    # Date and Time of the tweet
    tweet["created_at"] = status.created_at.strftime("%c")
    
    # Number of favorites in the tweet
    tweet["favorite_count"] = favorite
    
    #Number of retweets in the tweet
    tweet["retweet_count"] = status.retweet_count
    
    # Number of Photos in the tweet
    tweet["images_count"] = images_count
    
    tweets.append(tweet)

# Dump all the tweets in JSONline format into the file
fp.write(json.dumps(tweets))
fp.close()
```
